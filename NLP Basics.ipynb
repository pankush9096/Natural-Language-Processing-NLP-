### Text Preprocessing before BOW and TF-IDF transformation

```python
s= sent_tokenize(paragraph)
corpus=[]
for i in range(len(s)):
    r= re.sub('[^a-zA-Z]', ' ', s[i]) # replacing all the non char variable with space
    r= r.lower() # changing everythig to lower case
    r= word_tokenize(r) # tokenization to get words
    r= [lem.lemmatize(x) for x in r if x not in set(stopwords.words())] # lemmatizing all the actual words
    r= ' '.join(r) # creating back the sentence 
    corpus.append(r) # making new list with all new processed sentence
    
corpus
```

### Bag of words

```python
# SKlearn CountVectorizer will make histogram, then sort it out and will form bag of words matrix
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer()
x= cv.fit_transform(corpus).toarray()
x
```

### TF-IDF(Term frequency and inverse document frequency)

```python
#from sklearn we have module to get the tf-idf values of important words in paragraph which you see below
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf= TfidfVectorizer()
x_tfidf= tfidf.fit_transform(corpus).toarray()
x_tfidf
```
